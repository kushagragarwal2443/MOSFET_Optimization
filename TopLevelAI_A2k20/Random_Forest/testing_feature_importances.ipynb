# Important features according to RF: cqload, temp

```python
print(max(df_delay["cqload"]))
print(min(df_delay["cqload"]))
print(sum(df_delay["cqload"])/len(df_delay))

print(max(df_delay["temp"]))
print(min(df_delay["temp"]))
print(sum(df_delay["temp"])/len(df_delay))
```

```python
val_cq=2.28*(10**-15)
val_tm=69.62

for i in range(1, len(df_delay)):
    cqload_i = df_delay["cqload"][i]
    temp_i = df_delay["temp"][i]
    if(val_cq==cqload_i):
        if(abs(temp_i-val_tm)<10):
            print(i, abs(temp_i-val_tm))
```

```python
error1 = 0
error2 = 0

index1=0
index2=8831
index3=1561

for i in df_delay.columns:
    print(str(i[0:6])+"\t\t"+str(df_delay[i][index1])+"\t\t\t"+str(df_delay[i][index2])+"\t\t\t"+str(df_delay[i][index3]))
    error1+= abs(df_delay[i][index1]-df_delay[i][index2])/df_delay[i][index1]
    error2+= abs(df_delay[i][index1]-df_delay[i][index3])/df_delay[i][index1]
    
print("\n\n")
print((error1/16)*100, (error2/16)*100)
```
